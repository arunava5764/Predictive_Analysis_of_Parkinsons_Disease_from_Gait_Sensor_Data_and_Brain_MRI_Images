```python
p=['min', 'max', 'mean', 'median', 'std_dv', 'skew', 'kurt']
new_features=[]
k=features[1:]
for i in k:
    for j in p:
        new_features.append(i+"_"+j)
new_features.append("Label")
new_features.insert(0,"File_name")
print(new_features)
```

```python
h=[]
for name in os.listdir('CSV'):
    if 'Co' in name or 'Pt' in name:
        pd_df = pd.read_csv('CSV/' + name)
        l=[]
        l.append(name.split('.')[0])
        for i in k:
    #         trans_data[i+"_"+"min"]=pd_df[i].min()
    #         trans_data[i+"_"+"max"]=pd_df[i].max()
    #         trans_data[i+"_"+"mean"]=pd_df[i].mean()
    #         trans_data[i+"_"+"median"]=pd_df[i].median()
    #         trans_data[i+"_"+"std_dv"]=pd_df[i].std()
    #         trans_data[i+"_"+"skew"]=pd_df[i].skew()
    #         trans_data[i+"_"+"kurt"]=pd_df[i].kurt()
            l.append(pd_df[i].min())
            l.append(pd_df[i].max())
            l.append(pd_df[i].mean())
            l.append(pd_df[i].median())
            l.append(pd_df[i].std())
            l.append(pd_df[i].skew())
            l.append(pd_df[i].kurt())
        if 'Co' in name:
            l.append(0)
        elif 'Pt'in name:
            l.append(1)
        h.append(l)
trans_data = pd.DataFrame(h,columns = new_features)
print(trans_data) 
```
